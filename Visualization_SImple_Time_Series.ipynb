```

data_summary = df1.groupby(['SOURCE_KEY','DATE']).agg(READINGS = ('TOTAL_YIELD','count'),
                                                  INV = ('SOURCE_KEY', max),
                                                  DATE = ('DATE',max))
```

```
data_summary
```

```
data_summary.head()
```

```
plt.figure(figsize=(18,9))
for inv in data_summary['INV'].unique():

    plt.plot(data_summary[data_summary['INV']==inv]['DATE'],
            data_summary[data_summary['INV']==inv]['READINGS'],
            marker='o',
            linestyle='',
            alpha=.5,
            ms=10,
            label=inv)

plt.legend()
plt.title('Readings vs. Date')
plt.xlabel('Date')
plt.ylabel('Number of Readings')
plt.show()
```
